```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
  k = 0.0001
  return k*x
t = [2003, 2005, 2009, 2010, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022] 
solve = odeint(dydt, 504400, t=t)
solve = solve.flatten()
%matplotlib inline 

v = [504400, 502800, 502535, 502015, 508066, 509098, 509719, 510152, 510020, 510439, 509735, 509420, 508573, 503216, 496447]

plt.plot(t, solve, label = r"x′=kx")
plt.plot(t, v)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
a=-100000
b=0.5
def dydt(x,t):
  return k(x)*x*0.00355
def k(x):
  return a/x+b
t = [2003, 2005, 2009, 2010, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
solve = odeint(dydt, 504400, t=t)
solve = solve.flatten()
%matplotlib inline 

v = [504400, 502800, 502535, 502015, 508066, 509098, 509719, 510152, 510020, 510439, 509735, 509420, 508573, 503216, 496447]

plt.plot(t, solve, label = r"x'=k(x)*x")
plt.plot(t, v)
plt.grid(True)
plt.legend()
plt.show()
```
