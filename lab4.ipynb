```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def F(x, y):
    k = 0.3
    return k * y

def euler(F, x, y0, n):
    y = [y0]
    for i in range(n - 1):
        xi = x[i]
        yi = y[i]
        yi1 = yi + h * F(yi, xi)
        y.append(yi1)
    return x, y

h = 0.2

x = np.arange(0, 10 + h, h) 
t = np.linspace(0, 10, 100)
x, y = euler(F, y0 = 1, x = x, n = len(x - 1))

solve = odeint(F, y0 = 1, t = t)
solve = solve.flatten()

plt.plot(x, y, label = r"Метод Эйлера")
plt.plot(t, solve, label = r"x'=k(x)*x")
plt.grid(True)
plt.legend()
plt.title('График')
plt.show()
```
