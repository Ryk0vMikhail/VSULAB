Номер 1 Уравнение: x′ =kx

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
  k=0.42
  return k*x

t=np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t,solve,label=r"x′=kx")
plt.grid(True)
plt.legend()
plt.show()
```

Номер 2 Уравнение: x′ =k(x)x

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def k(x):
  a=1
  b=1
  return a*x-b

def dydt(x,t):
  return k(x)*x

t=np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t,solve,label=r"x′=k(x)*x")
plt.grid(True)
plt.legend()
plt.show()
```

Номер 3 Уравнение: x′ =k(x)x-c

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def k(x):
  a=1
  b=1
  return a*x-b

def dydt(x,t):
  c=1
  return k(x)*x-c

t=np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t,solve,label=r"x′=k(x)*x−c")
plt.grid(True)
plt.legend()
plt.show()
```

Номер 4 Уравнение: x′ =k(x)x-c(x)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def k(x):
  a=1
  b=2
  return a*x-b

def c(x):
  a=5
  return a*x

def dydt(x,t):
  return k(x)*x-c(x)

t=np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t,solve,label=r"x′=k(x)∗x−c(x)")
plt.grid(True)
plt.legend()
plt.show()
```
